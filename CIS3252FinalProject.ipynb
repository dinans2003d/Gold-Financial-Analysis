# Data Cleaning

Data cleaning is a crucial step in ensuring the accuracy and reliability of our analysis. This section outlines the procedures used to prepare the dataset for further exploration and modeling.

## Addressing Missing Data

Given the large amount of missing data in certain columns, it was necessary to employ various strategies to handle these effectively:

### Economic Indicators
- **CPI (Consumer Price Index)** and **US Interest Rates**: These columns had a significant number of missing entries. Given their importance in economic analysis, a linear interpolation method was used to fill in the gaps, preserving the general trend and seasonality.

```python
df['CPI'] = df['CPI'].interpolate(method='linear')
df['us_rates_%'] = df['us_rates_%'].fillna(method='ffill')
```

###Market Indices
#####S&P 500 and NASDAQ: Each of these columns was missing 185 entries. Since these gaps represent a small portion of the dataset and removing these entries does not significantly impact the dataset's integrity, rows with missing values in these columns were dropped.

```python
df.dropna(subset=['sp500 open', 'nasdaq open'], inplace=True)
```

###To ensure data integrity, checks were performed to identify any anomalies, such as negative values in columns where only positive values make sense (e.g., volumes and prices).



```python
negative_checks = ['sp500 volume', 'nasdaq volume', 'oil volume', 'gold volume']
for col in negative_checks:
    if (df[col] < 0).any():
        df[col] = df[col].apply(abs)
```

###Data Type Conversions
####It is essential that all data columns are of the correct data type for analysis. Numeric columns mistakenly captured as strings were converted to floats.

```python
df['eur_usd'] = pd.to_numeric(df['eur_usd'], errors='coerce')
```

###Drop duplicated data

```python
df = df.drop_duplicates()
```

###Normalization/Standardization
####Given that we'll be conducting statistical tests and regression analysis, normalizing data to have a mean of zero and a standard deviation of one ensures that the scales of the variables do not distort the results.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['sp500 open', 'nasdaq open']] = scaler.fit_transform(df[['sp500 open', 'nasdaq open']])
```

###Creating New Features
####New features were created to provide more insights into the data, such as calculating daily returns for the S&P 500 and NASDAQ, which are more indicative of market movement than absolute price data.


```python
df['sp500 daily return'] = df['sp500 close'].pct_change() * 100
df['nasdaq daily return'] = df['nasdaq close'].pct_change() * 100
```

## Bonus. Linear Regression Model


```python
# 'CPI' and 'sp500 close' are already loaded into the dataframe
df.dropna(subset=['CPI', 'sp500 close'], inplace=True)
```

```python
import statsmodels.api as sm

# Adding a constant to the predictor variable
X = sm.add_constant(df['CPI'])  # adding a constant for the intercept
y = df['sp500 close']

# fit the regression model
model = sm.OLS(y, X).fit()
```

```python
print(model.summary())
```

#####The summary from the Ordinary Least Squares (OLS) regression shows a strong relationship between the Consumer Price Index (CPI) and S&P 500 close prices, with an R-squared value of 0.916 indicating that about 91.6% of the variability in the S&P 500 is explained by changes in the CPI. The positive coefficient for CPI suggests that increases in CPI are associated with increases in S&P 500 close prices. The model's F-statistic is highly significant, confirming the model's overall statistical significance. However, the Durbin-Watson statistic indicates potential positive autocorrelation in the residuals, suggesting that adjustments for autocorrelation may be needed for more accurate results. Additionally, the significant Jarque-Bera test indicates that the residuals do not follow a normal distribution, which could impact the reliability of the regression's standard errors and t-statistics.









```python
import matplotlib.pyplot as plt

# Scatter plot of the data points
plt.scatter(df['CPI'], df['sp500 close'], alpha=0.5, label='Actual Data')

# Plotting the regression line
plt.plot(df['CPI'], model.predict(X), color='red', label='Regression Line')

plt.title('Linear Regression: CPI vs S&P 500 Close Prices')
plt.xlabel('CPI')
plt.ylabel('S&P 500 Close Price')
plt.legend()
plt.show()
```
